```python
import pandas as pd
import numpy as np
```

```python
pop20 = pd.read_csv('pop20.csv',encoding='euc-kr')
pop20.head()
```

```python
from datetime import date
def printDayOfTheWeek (year, month, day):
    dayOfTheWeek = ["월요일", "화요일", "수요일", "목요일", "금요일", "토요일", "일요일"]
    return dayOfTheWeek[date(year, month, day).weekday()]
```

```python
pop20 = pd.read_csv('pop20.csv',encoding='euc-kr')
pop20.head()
```

```python
pop20['요일'] = pop20.apply(lambda x : printDayOfTheWeek(int(str(x['일자']).strip()[0] + str(x['일자']).strip()[1] + str(x['일자']).strip()[2] + str(x['일자']).strip()[3]), int(str(x['일자']).strip()[4] + str(x['일자']).strip()[5]), int(str(x['일자']).strip()[6] + str(x['일자']).strip()[7])),axis=1)
```

```python
pop20.shape
```

```python
def quarter(x):
    if x in [1,2,3]:
        return 5
    elif x in [4,5,6]:
        return 6
    else : return 7
quarter(4)
```

```python
pop20['월'] = pop20.apply(lambda x: int(str(x['일자']).strip()[4]+str(x['일자']).strip()[5]), axis=1)
pop20['분기'] = pop20.apply(lambda x: quarter(x['월']), axis=1 )
```

```python
pop20.columns
```

```python
pop20['성별연령'] = pop20.apply(lambda x: str(x['연령대(10세단위)']) + "_" + x['성별'], axis=1 )
```

```python
pop20 = pop20[['분기', '군구','시간(1시간단위)', '성별연령','요일','유동인구수']]
```

```python
atime = pop20.pivot_table(index = ['분기','군구'], columns ='시간(1시간단위)', values='유동인구수',aggfunc='mean')
aagesex = pop20.pivot_table(index = ['분기','군구'], columns = '성별연령', values='유동인구수',aggfunc='mean')
aday = pop20.pivot_table(index = ['분기','군구'], columns = '요일', values='유동인구수',aggfunc='mean')
```

```python
atime = atime.reset_index()
aagesex = aagesex.reset_index()
aday = aday.reset_index()
```

```python
poptimeage = pd.merge(atime,aagesex,on=['분기', '군구'], how = 'outer')
poptimeageday = pd.merge(poptimeage, aday, on=['분기', '군구'], how = 'outer')
poptimeageday
```

```python
poptimeageday.to_csv('pop2020.csv',encoding='euc-kr',index=False)
```
